```python
import pandas as pd

data_path = "/data/lujd/neoag_data/main_task/"
hla_seq_df = pd.read_csv(data_path+"HLA_sequence_dict.csv", index_col=0)
hla_list = hla_seq_df.HLA_name.to_list()
len(hla_list)
```

```python
hla_A_list = []
for hla in hla_list:
    if hla.find("HLA-A") != -1:
        hla_A_list.append(hla)
len(hla_A_list)
```

```python
train_df = pd.read_csv(data_path+"train_data_fold4.csv", index_col=0)
train_df.head()
```

```python
train_hlaA_df = train_df[train_df.HLA.isin(hla_A_list)]
len(train_hlaA_df)
```

```python
train_hlaA_df.head()
```

```python
train_hlaA_df.groupby(by='HLA').count().sort_values(by='peptide', ascending=False)
```

```python
train_hlaA_pos_df = train_hlaA_df[train_hlaA_df.label==1]
len(train_hlaA_pos_df)
```

```python
hla_seq_dict = pd.read_csv(
        data_path+"HLA_sequence_dict.csv",
        index_col=0
        ).set_index(["HLA_name"])["clip"].to_dict()
list(hla_seq_dict.keys())==hla_list
```

```python
hla_B_list = []
for hla in hla_list:
    if hla.find("HLA-B") != -1:
        hla_B_list.append(hla)
len(hla_B_list)
```

```python
train_hlaB_df = train_df[train_df.HLA.isin(hla_B_list)]
print(len(train_hlaB_df))
train_hlaB_pos_df = train_hlaB_df[train_hlaB_df.label==1]
print(len(train_hlaB_pos_df))
train_hlaB_df.groupby(by='HLA').count().sort_values(by='peptide', ascending=False)
```

```python
hla_C_list = []
for hla in hla_list:
    if hla.find("HLA-C") != -1:
        hla_C_list.append(hla)
len(hla_C_list)
```

```python
train_hlaC_df = train_df[train_df.HLA.isin(hla_C_list)]
print(len(train_hlaC_df))
train_hlaC_pos_df = train_hlaC_df[train_hlaC_df.label==1]
print(len(train_hlaC_pos_df))
train_hlaC_df.groupby(by='HLA').count().sort_values(by='peptide', ascending=False)
```
