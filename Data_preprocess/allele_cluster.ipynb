# Distance 

```python
import pandas as pd

data_path = "/data/lujd/neoag_data/main_task/"
hla_seq_df = pd.read_csv(data_path+"HLA_sequence_dict_ABCEG.csv", index_col=0)
hla_abc_df = hla_seq_df[:112]

hla_abc_list = hla_abc_df["HLA_name"].to_list()
hla_abc_clip_list = hla_abc_df["clip"].to_list()
hla_abc_clip_list = ["G"+seq if len(seq)==181 else seq for seq in hla_abc_clip_list]

hla_seq_dict = hla_seq_df.set_index(["HLA_name"])["clip"].to_dict()
target_hla_list = [   
                        "HLA-A*03:02", "HLA-A*11:02", "HLA-B*42:01", "HLA-B*08:02",
                        "HLA-G*01:01", "HLA-E*01:03",
                        "HLA-A*33:03", "HLA-A*34:02"
                    ]
```

```python
import numpy as np
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("Rostlab/prot_bert_bfd", do_lower_case=False)     # only use the tokenizer
hla_abc_tokens = []
for seq in hla_abc_clip_list:
    seq = ' '.join(seq)
    hla_abc_tokens.append(tokenizer.encode(seq))
hla_abc_tokens = np.array(hla_abc_tokens)
hla_abc_tokens.shape
```

```python
target_hla = target_hla_list[-2]
hla_new_allele_token = np.array(
    [tokenizer.encode(
    ' '.join(hla_seq_dict[target_hla]))])
target_hla, hla_new_allele_token.shape
```

```python
def cal_Manhattan_dist(x, y):
    # given 1-D x, calculate Manhattan distance between x and y(1-D or 2-D)
    # x, y:  np.array, int
    if len(y.shape) == 1:
        return np.sum( (x!=y), axis=0 )
    elif len(y.shape) == 2:
        X = np.repeat(x.reshape(1,-1), y.shape[0], axis=0)
        return np.sum( (X!=y), axis=1 )
    
M_distances = cal_Manhattan_dist(hla_new_allele_token, hla_abc_tokens)

kdist = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
print("distance\t", kdist[:5], "\t", kdist[5:])

num_cluster = []
for i in kdist:
    num_cluster.append( sum(M_distances<i) )
print(target_hla, "\t", num_cluster[:5], "\t", num_cluster[5:])
```

```python
clusterset = []
for i, hla in enumerate(hla_abc_list):
    if M_distances[i] < 21:
        clusterset.append(hla)
clusterset = set(clusterset)
len(clusterset), clusterset
```

### Semantic Similarity 

```python
import pandas as pd

data_path = "/data/lujd/neoag_data/main_task/"
hla_seq_df = pd.read_csv(data_path+"HLA_sequence_dict_ABCEG.csv", index_col=0)
hla_abc_df = hla_seq_df[:112]

hla_abc_list = hla_abc_df["HLA_name"].to_list()
hla_abc_clip_list = hla_abc_df["clip"].to_list()
hla_abc_clip_list = ["G"+seq if len(seq)==181 else seq for seq in hla_abc_clip_list]

hla_seq_dict = hla_seq_df.set_index(["HLA_name"])["clip"].to_dict()
target_hla_list = [   
                        "HLA-A*03:02", "HLA-A*11:02", "HLA-B*42:01", "HLA-B*08:02",
                        "HLA-G*01:01", "HLA-E*01:03",
                        "HLA-A*33:03", "HLA-A*34:02"
                    ]
```

```python
import torch
import numpy as np
from tape import ProteinBertModel, TAPETokenizer

model = ProteinBertModel.from_pretrained('bert-base')
tokenizer = TAPETokenizer(vocab='iupac') 
model = model.cuda(7)
model.eval()

hla_abc_tokens = []
for seq in hla_abc_clip_list:
    hla_abc_tokens.append(tokenizer.encode(seq))
hla_abc_tokens = np.array(hla_abc_tokens)
hla_abc_tokens.shape
```

```python
target_hla = target_hla_list[7]
hla_new_allele_token = np.array([tokenizer.encode(hla_seq_dict[target_hla])])
target_hla, hla_new_allele_token.shape
```

```python
abc_tokens = torch.LongTensor(hla_abc_tokens).cuda(7)
new_token = torch.LongTensor(hla_new_allele_token).cuda(7)
abc_tokens.shape, new_token.shape
```

```python
with torch.no_grad():
    abc_repr = model(abc_tokens)[0]             # (112, 184, 1024)
    abc_repr = torch.mean(abc_repr, dim=1)
    new_repr = model(new_token)[0]
    new_repr = torch.mean(new_repr, dim=1)
abc_repr.shape, new_repr.shape
```

```python
abc_repr = abc_repr.cpu().detach().numpy()
new_repr = new_repr.cpu().detach().numpy()
abc_repr.shape, new_repr.shape
```

```python
import numpy as np

def cal_Euclidean_dist(x, y):
    if len(y.shape) == 1:
        return np.sum( np.square(x-y), axis=0 )
    elif len(y.shape) == 2:
        X = np.repeat(x.reshape(1,-1), y.shape[0], axis=0)
        return np.sum( np.square(X-y), axis=1 )
    
E_distances = cal_Euclidean_dist(new_repr, abc_repr)

kdist = np.arange(10)
print("distance\t", kdist[:5], "\t\t", kdist[5:])

num_cluster = []
for i in kdist:
    num_cluster.append( sum(E_distances<i) )
print(target_hla, "\t", num_cluster[:5], "\t", num_cluster[5:])
```

```python
clusterset = []
for i, hla in enumerate(hla_abc_list):
    if E_distances[i] < 8.45:
        clusterset.append(hla)
clusterset = set(clusterset)
len(clusterset), clusterset
```
