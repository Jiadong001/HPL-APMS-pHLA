```python
from collections import defaultdict

lengths = [8, 9, 10, 11, 12, 13, 14, 15]
HLA_list = list(set(remain_new_data['HLA']))
print(len(HLA_list))
    
pos_pep_segs_dict = defaultdict(set)
for HLA in sorted(HLA_list):
    cur_HLA_remain_new_data = remain_new_data[remain_new_data['HLA'] == HLA]
    curr_pep_set = set(cur_HLA_remain_new_data['peptide'].to_list())
    
    for length in lengths:
        for pep in curr_pep_set:
            if len(pep) < length:
                continue
            startpos, stoppos = 0, length
            while stoppos < len(pep) + 1:
                pos_pep_segs_dict[HLA].add(pep[startpos: stoppos])
                startpos += 1
                stoppos += 1
    print("HLA: {}\tpositive samples: {}\tsegment num: {}".format(HLA, len(curr_pep_set), len(pos_pep_segs_dict[HLA])))  
```

```python
# validation
for hla, pep_set in pos_pep_segs_dict.items():
    print(hla, Counter([len(_) for _ in pep_set]))

    for kk, vv in Counter([len(_) for _ in pep_set]).items():
        assert kk in set(list(range(8,16)))

    for pep in set(remain_new_data[remain_new_data['HLA']==hla]['peptide'].to_list()):
        assert pep in pep_set
        for length in range(8, 16):
            startpos, stoppos = 0, length
            while stoppos < len(pep) + 1:
                assert pep[startpos: stoppos] in pep_set
                startpos += 1
                stoppos += 1
```

```python
import numpy as np 

zeroshot_candidate_pool = defaultdict(set)
print(len(set(neg_pep_seg_pools)))

for HLA in remain_new_data.HLA.unique():
    zeroshot_candidate_pool[HLA] = set(neg_pep_seg_pools) - pos_pep_segs_dict[HLA]
for k, v in zeroshot_candidate_pool.items():
    print(k, len(v))
```

```python
def get_hla_seqs(fasta_name, hla_name):
    hla_seq_file = open(fasta_name,"r")
    line = hla_seq_file.readline()

    while line:
        if line.find(hla_name[4:]+":01:01") != -1 or line.find(hla_name[4:]) != -1:      # (HLA-)B*42:01 (+:01:01)
            seq_len = int(line[
                line.find(" bp")-3 : line.find(" bp")
                ])
            line = hla_seq_file.readline()

            seq = ""
            while line.find("bp") == -1:        # not find bp
                seq += line.strip("\n")
                line = hla_seq_file.readline()
            
            assert len(seq)==seq_len
            return seq
        else:
            line = hla_seq_file.readline()
```

```python
hla2seq_dict = {}
for hla in new_unique_hla_list:
    hla2seq_dict[hla] = get_hla_seqs("../neoag_data/raw_data/hla_prot.fasta", hla)
    print(hla, "length: ", len(hla2seq_dict[hla]))
```

```python
hla2seq_dict
```

```python
short_position = [7,9,24,45,59,62,63,66,67,69,70,73,74,76,77,80,81,84,95,97,99,
                    114,116,118,143,147,150,152,156,158,159,163,167,171]
HLA_list = []
clip_list = []
short_list = []
for hla, whole_seq in hla2seq_dict.items():
    HLA_list.append(hla)
    
    if len(whole_seq) < 182:
        clip_seq = whole_seq
    else:
        clip_seq = whole_seq[24:206]
    clip_list.append(clip_seq)
    
    if len(whole_seq) < 182:
        position = [pos-2 for pos in short_position]
    else:
        position = [pos-1 for pos in short_position]
    short_seq = ""
    for ind in position:
        short_seq += clip_seq[ind]
    short_list.append(short_seq)

hla_new_ab_seq_df = pd.DataFrame(columns=["HLA_name","full","clip","short"])
hla_new_ab_seq_df["HLA_name"] = HLA_list
hla_new_ab_seq_df["clip"] = clip_list
hla_new_ab_seq_df["short"] = short_list
hla_new_ab_seq_df
```
